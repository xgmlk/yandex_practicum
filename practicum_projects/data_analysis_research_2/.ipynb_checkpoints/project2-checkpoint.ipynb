# Исследование объявлений о продаже квартир
## Описание проекта  
В нашем распоряжении данные сервиса Яндекс Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости  
По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма.  
## Цель  
Научиться определять рыночную стоимость объектов недвижимости. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 
## Описание данных  
* `airports_nearest` — расстояние до ближайшего аэропорта в метрах (м)
* `balcony` — число балконов
* `ceiling_height` — высота потолков (м)
* `cityCenters_nearest` — расстояние до центра города (м)
* `days_exposition` — сколько дней было размещено объявление (от публикации до снятия)
* `first_day_exposition` — дата публикации
* `floor` — этаж
* `floors_total` — всего этажей в доме
* `is_apartment` — апартаменты (булев тип) P.S. это нежилые помещения, не относящиеся к жилому фонду, но имеющие необходимые условия для проживания.
* `kitchen_area` — площадь кухни в квадратных метрах (м²)
* `last_price` — цена на момент снятия с публикации
* `living_area` — жилая площадь в квадратных метрах (м²)
* `locality_name` — название населённого пункта
* `open_plan` — свободная планировка (булев тип)
* `parks_around3000`— число парков в радиусе 3 км
* `parks_nearest` — расстояние до ближайшего парка (м)
* `ponds_around3000` — число водоёмов в радиусе 3 км
* `ponds_nearest` — расстояние до ближайшего водоёма (м)
* `rooms` — число комнат
* `studio` — квартира-студия (булев тип)
* `total_area` — площадь квартиры в квадратных метрах (м²)
* `total_images` — число фотографий квартиры в объявлении

## Предобработка данных

### Обработка пропущенных значений

```python
df['is_apartment'] = df['is_apartment'].fillna(False) #Заполняем столбец "Аппатратменты"
df['days_exposition'] = df['days_exposition'].fillna(df['days_exposition'].median()) #Заполняем дни с момента публикации
df['floors_total'] = df['floors_total'].fillna(df['floor']) #Заполняем общее кол-во этажей
df['living_area'] = df['living_area'].fillna(df['total_area'] / 100 * 57) #Заполняем столбец "жилая площадь" отняв 57% от общей площади помещения
df['balcony'] = df['balcony'].fillna(0) #Заполняем пропущенные значения balcony
df['ceiling_height'] = df['ceiling_height'].fillna(df['ceiling_height'].median()) #Заменяем на медианное значение
```

```python
df[['living_area', 'floors_total','days_exposition', 'is_apartment', 'ceiling_height']].head(5)
```

### Замена типов данных

```python
df['last_price'] = df['last_price'].astype('int') #для удобства заменяем на целое число
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S') #устанавливаем правильный формат для дат
df['floors_total'] = df['floors_total'].astype('int') # так же для удобства вычислений устанавливает формат целое число
df['balcony'] = df['balcony'].astype('int') #целое число для удобства
df['days_exposition'] = pd.to_numeric(df['days_exposition'], errors='coerce', downcast='integer') # еще один вариант для сокращения занимаемого места в памяти
df['is_apartment'] = df['is_apartment'].astype('bool')
df[['last_price', 'first_day_exposition', 'floors_total', 'balcony', 'days_exposition' , 'is_apartment']].dtypes
```

## Исследовательский анализ данных

### Добавим столбцы
* Цена квадратного метра
* День недели, месяц и год публикации объявления
* Категория этажа (первый, последний, другой)
* соотношение жилой и общей площади, отношение площади кухни к общей

```python
df['avg_price_per_metr'] = df['last_price'] / df['total_area'] # Стобец цена квадратного метра
```

```python
df['dayp'] = df['first_day_exposition'].dt.weekday # День
df['monthp'] = df['first_day_exposition'].dt.month #Месяц
df['yearp'] = df['first_day_exposition'].dt.year # Год
```

```python
def floor_category(row): # Функция категоризации этажа по строкам
    if row['floor'] == 1:
        return 'Первый'
    if row['floor'] == row['floors_total']:
        return 'Последний'
    else:
        return 'Другой'

df['floor_category'] = df.apply(floor_category, axis = 1)
```

```python
df['ratio_living_area'] = df['living_area'] / df['total_area'] # Соотношение жилой площади к общей
df['ratio_kitchen_area'] = df['kitchen_area'] / df['total_area'] # Соотношение кухни к общей площади
```
